```python
df_hsinchu_population = df.query("`區域別` == '新竹市' and `性別` == '男女合計'")
df_hsinchu_population = df_hsinchu_population.loc[:, ['年月', '人口數']]

# 將年月資料型態變更為str, 使x軸表現更連續 (原本10412會直接跳10501)
df_hsinchu_population = df_hsinchu_population.astype({"年月": str})
df_hsinchu_population = df_hsinchu_population.set_index('年月')

df_hsinchu_population.plot(kind='line', figsize=(10, 6), fontsize=16)
plt.title('新竹市人口變化圖')
plt.show()
print(df_hsinchu_population.to_csv('新竹人口變化.csv'))
```

```python
#從上圖發現，111年是呈現負成長；因此最後針對111年做每月人口變化圖查看
df['年月']  = df['年月'].astype('str')  #年月的type為數值，須轉為str
df['年月']
```

```python
df_latest  = df.query("`年月`.str.contains('111') and `性別` == '男女合計'").groupby(['年月']).sum()
df_latest = df_latest.loc[:, ['人口數']]
df_latest
```

```python
#從11101開始每個月往下降低，直到11105開始有往上的趨勢
df_latest.plot(kind='line', figsize=(10, 8), fontsize=10)
plt.title('111年人口變化圖')
plt.show()
```
