```python
df_age_f = df_age.set_index('年紀')
df_age_f = df_age_f.loc[:,['高投入人數','低投入人數']]
df_age_f
```

```python
#逐步分析其他「學生背景」因素對於學習投入的影響？
#學科背景分佈
df_student_major = pd.read_csv('student_ major.csv')
df_student_major = df_student_major.set_index('學科背景')
df_student_major = df_student_major.loc[:,['高投入人數','低投入人數']]
df_student_major
```

```python
mx1 = df_student_major.index
my1 = df_student_major['高投入人數'].values
my2 = df_student_major['低投入人數'].values
```

```python
fig, axe = plt.subplots(figsize = (10, 6))
labels = mx1
values = my1
axe.pie(values, labels=labels,autopct='%.2f%%')
plt.title('高投入學生背景')
plt.show() #從圖表上發現Group2 (理工組)的背景學生投入程度會較高
```

```python
fig, axe = plt.subplots(figsize = (10, 6))
labels = mx1
values = my2
axe.pie(values, labels=labels,autopct='%.2f%%')
plt.title('低投入學生背景')
plt.show() #從圖表上發現Group1 (文科)的背景學生投入程度會較低
```

```python
#逐步分析其他「學生背景」因素對於學習投入的影響？
#學習目標分佈
df_student_target = pd.read_csv('student_target.csv')
df_student_target = df_student_target.set_index('學習目標')
df_student_target = df_student_target.loc[:,['高投入人數','低投入人數']]
df_student_target
```

```python
tx1 = df_student_target.index
ty1 = df_student_target['高投入人數'].values
ty2 = df_student_target['低投入人數'].values
```

```python
fig, axe = plt.subplots(figsize = (10, 6))
labels = tx1
values = ty1
axe.pie(values, labels=labels,autopct='%.2f%%')
plt.title('高投入學生背景')
plt.show() #從圖表上發現目標為Career Switcher 轉職者投入程度會較高
```

```python
fig, axe = plt.subplots(figsize = (10, 6))
labels = tx1
values = ty2
axe.pie(values, labels=labels,autopct='%.2f%%')
plt.title('低投入學生背景')
plt.show() #從圖表上發現目標為旅行者的低投入程度較高
```

```python
df_student_language = pd.read_csv('student_language.csv')
df_student_language = df_student_language.set_index('學過幾種程式')
df_student_language = df_student_language.loc[:,['高投入人數','低投入人數']]
df_student_language
```

```python
lauguage_x1 = df_student_language.index
lauguage_y1 = df_student_language['高投入人數'].values
lauguage_y2= df_student_language['低投入人數'].values
lauguage_y2
```

```python
fig, axe = plt.subplots(figsize = (12, 8))
labels = lauguage_x1
values = lauguage_y1
axe.pie(values, labels=labels,autopct='%.2f%%')
plt.title('高投入學生背景')
plt.legend()
plt.show() #從圖表上發現目標為學習過2-3種以上程式語言的學生，投入程度會較高
```

```python
fig, axe = plt.subplots(figsize = (12, 8))
labels = lauguage_x1
values = lauguage_y2
axe.pie(values, labels=labels,autopct='%.2f%%')
plt.title('低投入學生背景')
plt.legend()
plt.show() #從圖表上發現發現目標為學習過0-1以上程式語言的學生，投入程度會較低
```
